```python
from tensorflow.keras.models import Sequential #keep adding layers one after another
import matplotlib.pyplot as plt
import random
import numpy as np
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
```

```python
#load dataset
X_train = np.loadtxt('input.csv', delimiter = ',') #pixel values of images
Y_train = np.loadtxt('labels.csv', delimiter = ',')
X_test = np.loadtxt('input_test.csv', delimiter = ',')
Y_test = np.loadtxt('labels_test.csv', delimiter = ',')
```

```python
X_train = X_train.reshape(len(X_train), 100, 100, 3)
Y_train = Y_train.reshape(len(Y_train), 1)
#CNN expects input as (batch_size, height, width, channels) not as a flat row.
#Before reshaping → X_train is 2D
#After reshaping → it becomes 4D
X_test = X_test.reshape(len(X_test), 100, 100, 3)
Y_test = Y_test.reshape(len(Y_test), 1)
#normalization
#Pixel values in images are usually 0–255 (integers).
#Dividing by 255.0 scales them to 0–1 range (floating point)
X_train = X_train/255.0
X_test = X_test/255.0
```

```python
print("Shape of X_train:", X_train.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_test:", Y_test.shape)
```

```python
#visualizing a random training image
idx=random.randint(0,len(X_train))
plt.imshow(X_train[idx,:])#prepare this image to be displayed
plt.show()
```

```python
#Model
model=Sequential([
    Conv2D(32,(3,3),activation='relu',input_shape = (100, 100, 3)),#3,3 height and width of filter
    MaxPooling2D((2,2)), #32- number of filters and 3,3 size of the filter
    
    Conv2D(32, (3,3), activation = 'relu'),
    MaxPooling2D((2,2)),
    
    Flatten(), # convert a multi-dimensional array into a one-dimensional array. 
    Dense(64, activation = 'relu'), #dense - fully connected layers 64- neurons
    Dense(1, activation = 'sigmoid') # output layer 1 neuron - sigmoid - cause its binary classification
])
```

```python
model.compile(loss='binary_crossentropy', optimizer='SGD', metrics=['accuracy'])#to compile model by adding loss and backpropogation
#Optimizer = the algorithm that updates weights during training. updates weights step by step after each batch.
```

```python
H = model.fit(X_train, Y_train, epochs = 15, batch_size = 64)
#The training data is divided into batches of 64 images, updating the weights after every 64 images
```

```python
model.evaluate(X_test, Y_test)#Test the trained model on unseen test data.
```

```python
idx2 = random.randint(0, len(Y_test))
plt.imshow(X_test[idx2, :])
plt.show()

y_pred = model.predict(X_test[idx2, :].reshape(1, 100, 100, 3))
#gives a prediction probability.
#Example: [[0.82]] means 82% chance it is a cat.
y_pred = y_pred > 0.5#This converts probability into binary classification:

if(y_pred == 0):
    pred = 'dog'
else:
    pred = 'cat'
    
print("Our model says it is a :", pred)
```

```python
plt.plot(H.history['accuracy'], label='Training Accuracy')
plt.plot(H.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training vs Validation Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
